```python
# EDA
# 1. Number of earthquakes by year
# convert time to a datetime dtype
df['time'] = pd.to_datetime(df.time)

# create new columns for year and month
df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month

# visualize annual earthquakes
annual_eqs = df['time'].dt.year.value_counts().sort_index()
_ = annual_eqs.plot()
_ = plt.xlabel('year')
_ = plt.ylabel('earthquakes')
plt.show()
```

```python
# 2. Magnitude by year
# visualize average magnitude
_ = sns.lineplot(data=df, x='year', y='mag', ci=None)
_.set_ylim(0,4)
plt.show()
```
